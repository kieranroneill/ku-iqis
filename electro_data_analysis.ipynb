## 2. Pair probability rate

```python
datasets = [
    zero_attenuation_data,
    low_attenuation_data,
    medium_attenuation_data,
    high_attenuation_data
]
pump_repetition_rate = 1.23e9
total_photon_count = zero_attenuation_data[zero_attenuation_data.columns[1]].sum()
probabilities = [libs.calculate_pair_probability_rate(data=data, total_photon_count=total_photon_count) for data in datasets]
attenuation = range(len(datasets))

plt.plot(attenuation, probabilities, marker="o")
plt.xticks(attenuation, ["0", "4", "7", "10"])
plt.ylabel("Pair production probability per pump")
plt.xlabel("Attenuation of pump (dB)")
plt.title("Probability of generating at least one photon pair as a function of pump attenuation")
plt.show()
```
